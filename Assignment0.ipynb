```python
text_dict={}
for words in text_list:
    if words in text_dict:
        text_dict[words] += 1
    else:
        text_dict[words] = 1

stop_words = ["the","and","of","a","in",'to','The','is','are','as','on','that','with','for','not','from','has','have','by','at','be','its','would','an']
for i in stop_words:
    text_dict.pop(i)
```

```python
text_rank=sorted(text_dict.items(),key=lambda item:item[1],reverse = True)
text_rank[0:15]
```

```python
import csv
with open('keyword frequency.csv','w',newline='') as f:
    writer=csv.writer(f)
    header=['keyword','frequency']
    writer.writerow(header)
    writer.writerows(text_rank[0:15])
```

```python
#函数function
def extract(filename):
    text_list=[]
    delset=string.punctuation
    delset += '”“'
    for line in open(filename):
        if line.strip('“”') != '':
            line=line.translate(str.maketrans('','',delset))
            line=line.split()
            text_list.extend(line)
    text_dict={}
    for words in text_list:
        if words in text_dict:
            text_dict[words] += 1
        else:
            text_dict[words] = 1
    stop_words = ["the","and","of","in",'to','The','is','are','as','on','that','with','for','not','from','has','have','by','at','a','its','would','an']
    for i in stop_words:
        if i in text_dict:
            text_dict.pop(i)
    text_rank=sorted(text_dict.items(),key=lambda item:item[1],reverse = True)
    text_rank[0:15]
    return text_rank
```

```python
text_list=extract('assignment0/trade-wars-news1.txt')
text_list += extract('assignment0/trade-wars-news2.txt')
text_list += extract('assignment0/trade-wars-news3.txt')
text_list += extract('assignment0/trade-wars-news4.txt')
text_list += extract('assignment0/trade-wars-news5.txt')
text_list[0:15]
```

```python
import pandas as pd
df = pd.DataFrame(text_list[0:15])
df.columns.values
df.rename(columns={ df.columns[0]: "keyword", df.columns[1]: "frequency"}, inplace=True)
df.columns.values
df.to_csv('keywords frequency2.csv')
```
