```python
#以下一共爬了5个月份

from selenium import webdriver
import time


url = 'http://www.yue365.com/movie/time/201811.shtml'

    
browser = webdriver.Chrome()
browser.get(url)
time.sleep(2)

def get_movieinfo_from_browser(b):
    movie = []
    for session in browser.find_elements_by_css_selector('#tvli .cont_20'):
        data = {}
        data['Name'] = session.find_element_by_css_selector(".TV_name a").text
        data['Actors'] = session.find_element_by_css_selector("span.p").text
        data['Time'] = session.find_element_by_xpath("//*[@id='tvli']/li[1]/div[2]/ul/li[1]").text
        data['Intro'] = session.find_element_by_css_selector("span.p2").text
        movie.append(data)
    for session in browser.find_elements_by_css_selector('#tvli .cont_40'):
        data = {}
        data['Name'] = session.find_element_by_css_selector(".TV_name a").text
        data['Actors'] = session.find_element_by_css_selector("span.p").text
        data['Time'] = session.find_element_by_xpath("//*[@id='tvli']/li[1]/div[2]/ul/li[1]").text
        data['Intro'] = session.find_element_by_css_selector("span.p2").text
        movie.append(data)
    return movie

all_page_movieinfo11 = []
for i in range(1):
    time.sleep(0.5)
    try:
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight/1.5);')
        more_page = browser.find_element_by_xpath("//*[@id='addtv']/span")
        more_page.click()
        new_movieinfo11 = get_movieinfo_from_browser(browser)
        all_page_movieinfo11.extend(new_movieinfo11)

        
    except Exception as e:
        print(e)
        print('Error on page %s' % i)


#import pandas as pd 
#df = pd.DataFrame(all_page_movieinfo11)
#df
```
